The Sleep Health and Lifestyle Dataset comprises 400 rows and 13 columns, covering a wide range of variables related to sleep and daily habits. Detailed variable explanations are as follows:
(Dataset Column)

*   Person ID: An identifier for each individual.
*   Gender: The gender of the person (Male/Female).
*   Age: The age of the person in years.
*   Occupation: The occupation or profession of the person.
*   Sleep Duration (hours): The number of hours the person sleeps per day.
*   Quality of Sleep (scale: 1-10): A subjective rating of the quality of sleep, ranging from 1 to 10.
*   Physical Activity Level (minutes/day): The number of minutes the person engages in physical activity daily.
*   Stress Level (scale: 1-10): A subjective rating of the stress level experienced by the person, ranging from 1 to 10.
*   BMI Category: The BMI category of the person (e.g., Underweight, Normal, Overweight).
*   Blood Pressure (systolic/diastolic): The blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure.
*   Heart Rate (bpm): The resting heart rate of the person in beats per minute.
*   Daily Steps: The number of steps the person takes per day.
*   Sleep Disorder: The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea).

## 3.1 Pie chat

Make a pie chart to view the distribution of sleep disorders, occupation, and BMI indicators in the data.


```python
# pie chart
# Sleep Disorder
plt.figure(figsize = (6,6))
data['Sleep Disorder'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Sleep Disorder')
plt.show()
# Occupation
plt.figure(figsize = (8,8))
data['Occupation'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Occupation')
plt.show()
#BMI Category
plt.figure(figsize = (6,6))
data['BMI Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('BMI Category')
plt.show()
```

## 3.2 Bar chart

Create a bar chart to visually display sleep quality in different BMI categories, the distribution of BMI categories in different occupations, and the distribution of BMI categories in patients with sleep disorders. Explore the relationship between BMI categories and sleep quality, the impact of occupation on BMI indicators, and the impact of BMI classification on sleep disorders.

```python
# Bar chart
# Distribution of sleep quality and BMI
plt.figure(figsize=(10, 6))
bmi_counts = data.groupby('Quality of Sleep')['BMI Category'].value_counts().unstack()
bmi_counts.plot(kind='bar', stacked=True, colormap='viridis')
plt.xlabel('Quality of Sleep')
plt.ylabel('Count')
plt.title('Distribution of sleep quality and BMI')
plt.show()
# Occupation and BMI Category distribution
plt.figure(figsize=(10, 6))
bmi_counts = data.groupby('Occupation')['BMI Category'].value_counts().unstack()
bmi_counts.plot(kind='bar', stacked=True)
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.title('Occupation and BMI Category')
plt.show()
# Sleep Disorder and BMI Category
plt.figure(figsize=(10, 6))
bmi_counts = data.groupby('Sleep Disorder')['BMI Category'].value_counts().unstack()
bmi_counts.plot(kind='bar', stacked=True)
plt.xlabel('Sleep Disorder')
plt.ylabel('Count')
plt.title('Sleep Disorder and BMI Category')
plt.show()
```

As can be seen from the picture above, most people's sleep quality scores range from 6 to 9. The sleep scores of normal people are concentrated at 8 points; the scores of ordinary-weight people are concentrated at 8 and 9 points; the scores of obese people are concentrated at 7 and 4 points; the sleep quality indicators of overweight people appear in every segment, most of them Distributed in 6 points. It can be seen that people with normal BMI and normal weight tend to have higher sleep quality scores, while overweight people tend to have lower scores.


Most people in professions such as accountants, doctors, engineers, and lawyers have normal BMI indicators, while most people in professions such as nurses, salespeople, and teachers are overweight (obese and overweight).
Occupation may have some influence on people's BMI.

People with normal and normal weight in the BMI Category are more likely to have no sleep disorders, while overweight people are concentrated in insomnia and sleep apnea. Therefore, we suspect that people who are overweight are somehow linked to sleep disorders such as insomnia or sleep apnea, and that being overweight may be a predisposing cause of sleep disorders.

## 3.3 Box plot

```python
# Box plot
# Sleep duration group by Sleep Disorder
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sleep Disorder', y='Sleep Duration', data=data, showfliers=False)
sns.swarmplot(x='Sleep Disorder', y='Sleep Duration', data=data, color=".25")
plt.title('Sleep duration group by Sleep Disorder')
plt.show()
```

As we know, adults need 7 to 9 hours of sleep per day.(https://www.nhlbi.nih.gov/health/sleep/how-much-sleep)
The sleep duration of people without sleep disorders is concentrated in 7 to 8 hours, which meets the daily required sleep time; the sleep duration of people with sleep apnea is widely distributed, between 6 and 8 hours, although some people are concentrated in 8 hours. However, more than half of the people cannot achieve the required daily sleep time; the sleep duration of people with insomnia is concentrated between 6 and 7 hours, which cannot meet the daily required sleep time.

```python
# Stress Level group by Sleep Disorder
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sleep Disorder', y='Stress Level', data=data, showfliers=False)
sns.swarmplot(x='Sleep Disorder', y='Stress Level', data=data, color=".25")
plt.title('Stress Level group by Sleep Disorder')
plt.show()
```

Judging from the results, the stress level of the general population is low, the pressure of sleep apnea patients is severely divided into two levels, and the stress level of insomnia patients is more distributed between level 7 and level 4. Combined with the distribution of sleep duration, it can be inferred that short sleep duration may lead to excessive stress and thus reduce sleep quality.

```python
# Sleep Duration group by BMI Category
plt.figure(figsize=(10, 6))
sns.boxplot(x='BMI Category', y='Sleep Duration', data=data, showfliers=False)
sns.swarmplot(x='BMI Category', y='Sleep Duration', data=data, color=".25")
plt.title('Sleep Duration group by BMI Category')
plt.show()
```

Overweight people generally have low sleep duration; people with normal and normal weight mostly get enough sleep every day, and obese people tend to have insufficient sleep duration.

## 3.4 Dot plot

Create a dot plot to analyze the relationship between sleep quality and sleep duration.

```python
# Dot plot
# The relationship between Quality of Sleep and sleep duration.
plt.figure(figsize=(12, 6))
sns.pointplot(x='Sleep Duration', y='Quality of Sleep', data=data, dodge=True)
plt.title('The relationship between Quality of Sleep and sleep duration')
plt.show()
```

From the dot plot, we can easily see that sleep quality becomes better as the sleep duration increases. Therefore, we can speculate that in order to obtain better sleep quality, we can try to increase the sleep time.

## 3.5 Scatterplot matrix

Create a scatter matrix plot of heart rate, sleep quality and stress indicators to explore the relationship between the four.

```python
# Relationships between heart rate, sleep quality, and stress indicators.
sns.pairplot(data[['Heart Rate', 'Quality of Sleep', 'Stress Level']], height=2.5)
plt.suptitle('Multivariable relationship diagram', y=1.02)
plt.show()
```

It can be seen from the multivariate relationship diagram that the heart rate in the sample is concentrated between 65 and 75, the sleep quality level is distributed between 6 and 9, and the stress level is approximately uniformly distributed.

At the same time, we can clearly see in the figure that stress levels are negatively correlated with sleep quality, heart rate is positively correlated with stress levels, and negatively correlated with sleep quality. This suggests that lowering stress levels may improve sleep quality and lower resting heart rate, and lowering heart rate may improve sleep quality.
